```python
##Section A :1.HealthCare Plus recorded the daily number of patient admissions for the past 10 days:
#[32, 28, 35, 30, 29, 27, 31, 34, 33, 30]
#Compute the mean, median, and mode of patient admissions.
#Which measure best represents patient admissions?
#If the hospital increases its admission capacity by 10%, how will this affect the measures of central tendency?

import numpy as np
from scipy import stats

# Patient admission data for the past 10 days
admissions = [32, 28, 35, 30, 29, 27, 31, 34, 33, 30]

# Calculate the mean, median, and mode
mean_admission = np.mean(admissions)
median_admission = np.median(admissions)
mode_result = stats.mode(admissions)

# Directly access mode and count
mode_admission = mode_result.mode  # The mode value (scalar)
mode_count = mode_result.count  # The count of the mode (scalar)

# Display the original measures of central tendency
print(f"Original Data: {admissions}")
print(f"Mean: {mean_admission}")
print(f"Median: {median_admission}")
print(f"Mode: {mode_admission} (appears {mode_count} times)" if mode_admission else "Mode: No mode found")

# Increase the admission capacity by 10%
increased_admissions = [admission * 1.10 for admission in admissions]

# Calculate the new mean, median, and mode after the increase
mean_increased = np.mean(increased_admissions)
median_increased = np.median(increased_admissions)
mode_result_increased = stats.mode(increased_admissions)

# Directly access new mode and count
mode_increased = mode_result_increased.mode  # The new mode value (scalar)
mode_count_increased = mode_result_increased.count  # The new count of the mode (scalar)

# Display the new measures of central tendency
print(f"\nData After 10% Increase: {increased_admissions}")
print(f"New Mean: {mean_increased}")
print(f"New Median: {median_increased}")
print(f"New Mode: {mode_increased} (appears {mode_count_increased} times)" if mode_increased else "New Mode: No mode found")

# Interpretation of results
print("\nInterpretation:")
print("1. The mean is affected by every value, so it will increase with the 10% rise in patient admissions.")
print("2. The median will change if the central value shifts due to the increased capacity.")
print("3. The mode may not change if the relative frequency of admissions remains the same after the increase.")
```

```python
##Section A :2. The recovery duration (in days) of 10 patients who underwent the same surgery is recorded as follows:
#[5, 7, 6, 8, 9, 5, 6, 7, 8, 6]
#Calculate the range, variance, and standard deviation.
#What does the standard deviation indicate about variability in recovery times?
#If two new patients take 4 and 10 days to recover, how will this impact the standard deviation?

import numpy as np

# Recovery duration data for 10 patients
recovery_times = [5, 7, 6, 8, 9, 5, 6, 7, 8, 6]

# Calculate the range (difference between max and min)
range_recovery = np.max(recovery_times) - np.min(recovery_times)

# Calculate variance
variance_recovery = np.var(recovery_times)

# Calculate standard deviation
std_dev_recovery = np.std(recovery_times)

# Display the original measures
print(f"Original Data: {recovery_times}")
print(f"Range: {range_recovery}")
print(f"Variance: {variance_recovery}")
print(f"Standard Deviation: {std_dev_recovery}")

# Adding two new recovery times (4 and 10 days)
new_recovery_times = recovery_times + [4, 10]

# Calculate new variance and standard deviation after adding the two new values
variance_new = np.var(new_recovery_times)
std_dev_new = np.std(new_recovery_times)

# Display the new measures
print(f"\nNew Data After Adding 4 and 10 Days: {new_recovery_times}")
print(f"New Variance: {variance_new}")
print(f"New Standard Deviation: {std_dev_new}")

# Interpretation of standard deviation
print("\nInterpretation:")
print("1. Standard deviation measures how spread out the recovery times are from the mean.")
print("2. A higher standard deviation means more variability in the recovery times.")
print("3. Adding the new recovery times of 4 and 10 days will increase the standard deviation because they are farther from the original mean, indicating more variability in recovery times.")
```

```python
##Section A :3. Patient satisfaction scores (on a scale of 1 to 10) collected from a hospital survey are:
#[8, 9, 7, 8, 10, 7, 9, 6, 10, 8, 7, 9]
#Compute skewness and kurtosis.
#Interpret the resultsâ€”does the data suggest a normal distribution?
#If the hospital implements a new customer service initiative, and satisfaction scores shift higher, what type of skewness change would you expect?

import numpy as np
from scipy.stats import skew, kurtosis

# Patient satisfaction scores
satisfaction_scores = [8, 9, 7, 8, 10, 7, 9, 6, 10, 8, 7, 9]

# Calculate skewness and kurtosis
skewness = skew(satisfaction_scores)
kurt = kurtosis(satisfaction_scores)

# Display the skewness and kurtosis values
print(f"Satisfaction Scores: {satisfaction_scores}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

# Interpretation
print("\nInterpretation:")
if skewness > 0:
    print("The data is right-skewed (positively skewed), meaning there are more values on the lower end of the scale.")
elif skewness < 0:
    print("The data is left-skewed (negatively skewed), meaning there are more values on the higher end of the scale.")
else:
    print("The data is approximately symmetric.")

if kurt > 0:
    print("The distribution has heavier tails (leptokurtic), suggesting more outliers.")
elif kurt < 0:
    print("The distribution has lighter tails (platykurtic), indicating fewer outliers.")
else:
    print("The distribution has a normal kurtosis (mesokurtic), meaning it follows a normal distribution shape.")

# Prediction of skewness change after the customer service initiative
print("\nPrediction of Skewness Change After Customer Service Initiative:")
print("If the customer satisfaction scores shift higher (i.e., more scores approach the higher end of the scale), we would expect the skewness to become more negative, indicating a left-skewed distribution as more patients would give higher ratings.")
```

```python
##Section A :4. HealthCare Plus wants to analyze the relationship between nurse staffing levels and patient recovery time. Data from 6 hospital departments is provided:
#Number of Nurses :Average Recovery Time (days) :10,8,12,7,15,6,18,5,20,4,22,3
#Compute the correlation coefficient between nurse staffing and patient recovery time.
#If the hospital increases the number of nurses by 5 per department, how will this affect the recovery time based on the trend?

import numpy as np
from scipy.stats import pearsonr

# Data: Number of Nurses and Average Recovery Time (days)
nurses = [10, 12, 15, 18, 20, 22]  # Number of nurses in each department
recovery_time = [8, 7, 6, 5, 4, 3]  # Average recovery time in days

# Calculate the Pearson correlation coefficient
corr_coefficient, _ = pearsonr(nurses, recovery_time)

# Display the correlation coefficient
print(f"Pearson Correlation Coefficient: {corr_coefficient}")

# Interpretation of correlation
if corr_coefficient > 0:
    print("There is a positive correlation between the number of nurses and the recovery time.")
elif corr_coefficient < 0:
    print("There is a negative correlation between the number of nurses and the recovery time.")
else:
    print("There is no correlation between the number of nurses and the recovery time.")

# Predict the effect of increasing the number of nurses by 5
increased_nurses = np.array(nurses) + 5
# Calculate the new predicted recovery times based on the trend (using the correlation trend)
# The recovery time is inversely related to the number of nurses, so we can assume a decrease in recovery time
# corresponding to an increase in nurses.

print(f"\nNew number of nurses per department: {increased_nurses}")
```

```python
#Section B -5. The hospital claims that the average patient wait time in the emergency department is 30 minutes. A sample of 10 patient wait times (in minutes) is recorded:
#[32, 29, 31, 34, 33, 27, 30, 28, 35, 26]
#Test whether the hospitalâ€™s claim is valid at a 5% significance level.
#State the null and alternative hypotheses.
#If the wait time significantly exceeds 30 minutes, what changes should the hospital implement to reduce waiting time?

import numpy as np
from scipy import stats

# Sample of 10 patient wait times in minutes
wait_times = [32, 29, 31, 34, 33, 27, 30, 28, 35, 26]

# Sample size
n = len(wait_times)

# Calculate sample mean and sample standard deviation
sample_mean = np.mean(wait_times)
sample_std = np.std(wait_times, ddof=1)  # Using ddof=1 for sample standard deviation

# Population mean (claimed average)
population_mean = 30

# Compute t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Find the critical t-value for a two-tailed test at 5% significance level
alpha = 0.05
critical_t = stats.t.ppf(1 - alpha/2, df)

# Compare the t-statistic with the critical t-value
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-statistic: {t_statistic}")
print(f"Critical t-value: {critical_t}")

# Decision: Reject or Fail to Reject Null Hypothesis
if abs(t_statistic) > critical_t:
    print("\nReject the null hypothesis. The average wait time is significantly different from 30 minutes.")
else:
    print("\nFail to reject the null hypothesis. There is no significant difference from 30 minutes.")
```

```python
#Section B -6. A survey was conducted on hospital cleanliness and patient satisfaction and data was collected:
#Perform an analysis to check whether hospital cleanliness and patient satisfaction are dependent.
#If cleanliness ratings improve, how do you expect the distribution of satisfied and unsatisfied patients to change

import matplotlib.pyplot as plt
import numpy as np

# Data: Cleanliness Ratings and Satisfaction Data
cleanliness = ['High', 'Medium', 'Low']
satisfied = [90, 60, 30]
unsatisfied = [10, 40, 70]

# Calculate proportions of satisfied and unsatisfied patients in each category
total_patients = np.array(satisfied) + np.array(unsatisfied)
proportion_satisfied = np.array(satisfied) / total_patients
proportion_unsatisfied = np.array(unsatisfied) / total_patients

# Print the proportions
for i in range(len(cleanliness)):
    print(f"Cleanliness: {cleanliness[i]}")
    print(f"Proportion Satisfied: {proportion_satisfied[i]:.2f}")
    print(f"Proportion Unsatisfied: {proportion_unsatisfied[i]:.2f}\n")

# Create a bar plot to visualize the proportions
bar_width = 0.35
index = np.arange(len(cleanliness))

fig, ax = plt.subplots()

bar1 = ax.bar(index, proportion_satisfied, bar_width, label='Satisfied')
bar2 = ax.bar(index + bar_width, proportion_unsatisfied, bar_width, label='Unsatisfied')

# Labeling
ax.set_xlabel('Cleanliness Rating')
ax.set_ylabel('Proportion of Patients')
ax.set_title('Proportion of Satisfied vs Unsatisfied Patients by Cleanliness')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(cleanliness)
ax.legend()

plt.show()
```

```python
#Section B -7. The hospital tested three different treatment methods (A, B, and C) for managing post-surgery pain. 
#The recovery durations (in days) under each treatment are provided.
#Conduct an analysis to check if there is a significant difference in recovery times among the treatment methods.
#State the null and alternative hypotheses.
#If the hospital introduces a new treatment (D), what data should be collected before concluding its effectiveness?

import numpy as np
from scipy.stats import f_oneway

# Recovery durations for each treatment
treatment_A = [5, 6, 7, 5, 6]
treatment_B = [8, 9, 7, 8, 10]
treatment_C = [4, 5, 6, 5, 4]

# Perform One-Way ANOVA
f_statistic, p_value = f_oneway(treatment_A, treatment_B, treatment_C)

# Display the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference in recovery times among the treatment methods.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference in recovery times among the treatment methods.")
```

```python
#Section B -8.The hospital administration time (in minutes) for 12 patients is recorded as:
#[12, 15, 14, 16, 18, 13, 14, 17, 15, 19, 16, 14]
#Analyze whether the administration times follow a normal distribution.
#Explain why this analysis is important in healthcare data.
#If emergency cases increase, how would you expect the distribution of administration times to change?

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Administration times (in minutes)
admin_times = [12, 15, 14, 16, 18, 13, 14, 17, 15, 19, 16, 14]

# Visual inspection - Histogram
plt.figure(figsize=(10, 6))
sns.histplot(admin_times, kde=True, bins=6)
plt.title("Histogram of Administration Times")
plt.xlabel("Administration Time (minutes)")
plt.ylabel("Frequency")
plt.show()

# Visual inspection - Q-Q plot
plt.figure(figsize=(10, 6))
stats.probplot(admin_times, dist="norm", plot=plt)
plt.title("Q-Q Plot of Administration Times")
plt.show()

# Perform the Shapiro-Wilk test for normality
shapiro_stat, shapiro_p_value = stats.shapiro(admin_times)
print(f"Shapiro-Wilk test statistic: {shapiro_stat}")
print(f"Shapiro-Wilk test p-value: {shapiro_p_value}")

# Interpretation of the Shapiro-Wilk test
if shapiro_p_value < 0.05:
    print("\nReject the null hypothesis: The data does not follow a normal distribution.")
else:
    print("\nFail to reject the null hypothesis: The data follows a normal distribution.")
```

```python
#Section B -9.The hospital is studying the distribution of patient arrival times in the emergency department. 
#Historical data suggests that emergency cases arrive at an average rate of 5 per hour.
#Model this scenario using an appropriate probability distribution.
#What is the probability that exactly 3 emergency cases will arrive in the next hour?
#If a major accident occurs in the city, how would this affect the probability distribution of emergency arrivals?

from scipy.stats import poisson

# Average rate (lambda) of emergency cases per hour
lambda_rate = 5

# Number of cases for which we are calculating the probability
k = 3

# Calculate the Poisson probability for exactly 3 cases
probability_3_cases = poisson.pmf(k, lambda_rate)

print(f"The probability of exactly 3 emergency cases arriving in the next hour is: {probability_3_cases:.4f}")
```

```python
# New average rate (lambda) of emergency cases per hour after the accident
lambda_rate_new = 10

# Calculate the Poisson probability for exactly 3 cases with the new lambda
probability_3_cases_new = poisson.pmf(k, lambda_rate_new)

print(f"With the new rate (lambda = 10), the probability of exactly 3 emergency cases arriving in the next hour is: {probability_3_cases_new:.4f}")
```

```python
#Section B -10.TThe number of surgeries performed per day in the hospital follows a specific distribution pattern. Historical data is provided.
#Identify and justify the type of probability distribution that best fits this data.
#Calculate the expected number of surgeries performed per day.
#If a new surgical team is hired, how will this affect the probability distribution of daily surgeries?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Data: Number of surgeries performed per day (hypothetical example)
surgeries_per_day = [0, 1, 2, 3, 4, 5]  # Number of surgeries performed
frequencies = [5, 12, 18, 22, 15, 8]    # Frequency of each number of surgeries

# Calculate the average rate (lambda) of surgeries per day
lambda_rate = np.sum(np.array(surgeries_per_day) * np.array(frequencies)) / np.sum(frequencies)

print(f"Average number of surgeries per day (lambda): {lambda_rate:.2f}")

# Model the number of surgeries per day using Poisson distribution
x_values = np.arange(0, 6)  # Values for surgeries (0, 1, 2, 3, 4, 5)
poisson_probs = poisson.pmf(x_values, lambda_rate)  # Poisson PMF for each value

# Plot the probability mass function of the Poisson distribution
plt.figure(figsize=(8, 6))
plt.bar(x_values, poisson_probs, width=0.4, align='center', alpha=0.6, color='b', label='Poisson Distribution')
plt.xlabel('Number of Surgeries per Day')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution: Surgeries per Day (lambda = {lambda_rate:.2f})')
plt.xticks(x_values)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Calculating the probability of exactly 3 surgeries per day
probability_3_surgeries = poisson.pmf(3, lambda_rate)
print(f"Probability of exactly 3 surgeries in a day: {probability_3_surgeries:.4f}")
```

```python

# Data: surgeries performed and corresponding frequencies
surgeries = [0, 1, 2, 3, 4, 5]
frequencies = [5, 12, 18, 22, 15, 8]

# Calculate the expected number of surgeries (mean)
total_surgeries = sum(s * f for s, f in zip(surgeries, frequencies))
total_frequency = sum(frequencies)

expected_surgeries = total_surgeries / total_frequency

print(f"Expected number of surgeries per day: {expected_surgeries:.2f}")
```
